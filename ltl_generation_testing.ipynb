```python
# import and setup gemini
import json
from dotenv import load_dotenv
import os
from google import genai

import dataset_manager
import ltl_generation

# api key must be stored inside a file called .env with the following contents: GEMINI_API_KEY=<your_key_here>
load_dotenv()
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))
```

```python
# load the dataset from a json file
dataset = dataset_manager.load_dataset("dataset_iter2.json")
```

```python
for entry in dataset:
    for nlp in entry["nlp"]:
        ltl = ltl_generation.generate_nlp2ltl(client, nlp)
        print("-------------------------")
        print(nlp)
        print(f"Result: {ltl}")
        print(f"Expected: {entry["ltl"]}")
```
